## （問2）エアコンとアイスクリームの移動平均の可視化
- スライダーを動かして、移動平均の期間を変化させるとグラフの形状はどうなるかな？

```python
# 単純移動平均 (Simple Moving Average; SMA) 
def viewSMA(month):
    offset=month//2
    data2=[data[0]]
    max_val = 0
    min_val = 10000
    for i in range(1,len(data)-month):
        x = data[i+offset][0]
        y1, y2 = 0, 0
        for j in range(month):
            y1 += data[i+j][1]
            y2 += data[i+j][2]
        y1 = y1 / month
        y2 = y2 / month
        max_val = max(max_val, y1, y2)
        min_val = min(min_val, y1, y2)
        data2.append( [ x, y1, y2 ] )
    df = pd.DataFrame(data2[1:],columns=data2[0])
    viewLineChart(df, min_val, max_val, 50)

month = widgets.IntSlider( value=12, min=2, max=24, step=1, description="移動平均期間" )
widgets.interact( viewSMA, month=month)
```
