## 前処理

```python
import random
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import ipywidgets as widgets

# 与えられた数字（0以上1以下）をもとに、到着間隔を計算して整数として返す関数
def calcInterval(num):
    CRF = [0.12, 0.26, 0.42, 0.64, 0.82, 0.90, 0.94, 0.94, 0.96, 1.00]      # 累積相対度数:Cumulative Relative Frequency
    for i in range(len(CRF)):
        if num < CRF[i] :
            return i

# 乱数リストをもとに、客ごとの「待ち」と「対応」の開始終了時刻を計算して2種類の辞書型リストを返す関数
def calcTimeData(resp_time, rand_nums):
    wait_time_data = [ dict(Category="待ち時間", Guest="1人目", Start=0, Finish=0) ]
    resp_time_data = [ dict(Category="対応時間", Guest="1人目", Start=0, Finish=resp_time) ]
    # 2人目以降の計算
    for i in range(len(rand_nums)):
        ws_time = wait_time_data[i]["Start"] + calcInterval(rand_nums[i])   # 到着時刻 = 前の客の到着時刻 + 到着間隔
        rs_time = resp_time_data[i]["Finish"]                               # （接客中に到着したとき）接客開始時刻 = 前の客の接客終了時刻
        if rs_time < ws_time : rs_time = ws_time                            # （待機中に到着したとき）接客開始時刻 = 到着時刻
        rf_time = rs_time + resp_time                                       # 接客終了時刻
        wait_time_data.append( dict(Category="待ち時間", Guest=str(i+2)+"人目", Start=ws_time, Finish=rs_time) )
        resp_time_data.append( dict(Category="対応時間", Guest=str(i+2)+"人目", Start=rs_time, Finish=rf_time) )
    return (resp_time_data, wait_time_data)

# 「待ち」と「対応」の開始終了時刻をもとに、最大待ち人数を計算して整数で返す関数
def getMaxWaitCount(resp_time_data, wait_time_data):
    max_wait_count = 0
    # 方針：各客の後ろに並ぶ人数を数えて、最大待ち人数を見つけ出す
    for i in range(len(wait_time_data)-1):
        j = i+1
        wait_count = 0
        # i人目の客の対応が終了するまでに、j人目の客が到着したかを判定して、待ち人数を数えていく
        while resp_time_data[i]["Finish"] > wait_time_data[j]["Start"] and j < len(wait_time_data)-1:
            wait_count += 1
            j += 1
            if max_wait_count < wait_count :
                max_wait_count = wait_count
    return max_wait_count

# 「待ち」の開始終了時刻をもとに、最大待ち時間を計算して整数で返す関数
def getMaxWaitTime(wait_time_data):
    max_wait_time = wait_time_data[0]["Finish"] - wait_time_data[0]["Start"]
    for i in range(1,len(wait_time_data)):
        wait_time = wait_time_data[i]["Finish"] - wait_time_data[i]["Start"]
        if max_wait_time < wait_time :
            max_wait_time = wait_time
    return max_wait_time

# 「待ち」と「対応」の開始終了時刻をもとに、ガントチャートを表示する関数
def viewGanttchart(resp_time_data, wait_time_data):
    data = resp_time_data
    data.extend(wait_time_data)
    df = pd.DataFrame(data)
    df["time"] = df["Finish"] - df["Start"]
    fig = px.bar(df, base="Start", x="time", y="Guest", color="Category", orientation="h")
    fig.update_layout(bargap=0.1, width=1200)
    fig.update_xaxes(title="分", dtick=1)
    fig.update_yaxes(title="", autorange="reversed")
    fig.show()

# 最大待ち人数のリストをもとに、ヒストグラムを表示する関数
def viewHistogram(data, scales):
    subtitle = list(map(lambda x: "来客人数："+str(x), scales))
    fig = make_subplots(rows=1, cols=len(scales), subplot_titles=subtitle, shared_yaxes=True)
    fig.update_layout(width=1200, height=400, showlegend=False)
    fig.update_xaxes(title_text="最大待ち人数", range=(0,30), dtick=5)
    fig.update_yaxes(title_text="回数", range=(0,60), dtick=5 )
    for i in range(len(scales)):
        df = pd.DataFrame(data[i])
        counts, bins = np.histogram(df[0], bins=range(0, 31, 1))
        trace = go.Bar(x=bins[:30], y=counts)
        fig.append_trace(trace, 1, i+1)
    fig.show()


# 問１で用いられた客の待ち状況をシミュレートする関数　（resp_time:客一人あたりの対応時間, scale:来客人数）
def Simulation_A(resp_time, scale, *args):
    rand_nums = [ random.random() for _ in range(scale-1) ] if args == () else args[0]
    (resp_time_data, wait_time_data) = calcTimeData(resp_time, rand_nums)
    max_wait_count = getMaxWaitCount(resp_time_data, wait_time_data)
    print("最大待ち人数は", max_wait_count, "人")
    max_wait_time = getMaxWaitTime(wait_time_data)
    print("最大待ち時間は", max_wait_time, "分")
    viewGanttchart(resp_time_data, wait_time_data)

# 問２・問３で用いられた最大待ち人数をシミュレートする関数
def Simulation_B(resp_time):
    scales = [10,20,30,40]
    data=[ [] for _ in range(len(scales)) ]
    for j in range(len(scales)) :
        for i in range(100):
            rand_nums = [ random.random() for _ in range(scales[j]) ]
            (resp_time_data, wait_time_data) = calcTimeData(resp_time, rand_nums)
            max_wait_count = getMaxWaitCount(resp_time_data, wait_time_data)
            data[j].append( max_wait_count )
    viewHistogram(data, scales)

# 表２と同じ条件で客の待ち状況をシミュレートする関数
def Question1():
    EXAMPLE_RAND_NUMS = [0.31, 0.66, 0.41, 0.11, 0.63, 0.43, 0.28, 0.55, 0.95]
    resp_time = 4
    scale = 10
    print("生成させた乱数が 0.95 のとき、到着間隔は",calcInterval(0.95),"分")
    Simulation_A(resp_time, scale, EXAMPLE_RAND_NUMS)
```
